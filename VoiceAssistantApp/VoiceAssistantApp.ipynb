```python
import speech_recognition as sr
import playsound
from gtts import gTTS
import os
import wolframalpha
from selenium import webdriver

# Initialize the WolframAlpha client with your API key
app_id = 'YOUR_WOLFRAM_ALPHA_API_KEY'  # Replace with your WolframAlpha API key
client = wolframalpha.Client(app_id)

num = 1

def assistant_speaks(output):
    global num

    # Increment num to rename every audio file
    num += 1
    print("PerSon : ", output)

    toSpeak = gTTS(text=output, lang='en', slow=False)
    file = str(num) + ".mp3"
    toSpeak.save(file)

    # Play the saved file
    playsound.playsound(file, True)
    os.remove(file)

def get_audio():
    rObject = sr.Recognizer()
    audio = ''

    with sr.Microphone() as source:
        print("Speak...")
        audio = rObject.listen(source, phrase_time_limit=5)
    print("Stop.")  # Limit 5 secs

    try:
        text = rObject.recognize_google(audio, language='en-US')
        print("You : ", text)
        return text

    except sr.UnknownValueError:
        assistant_speaks("Could not understand your audio, please try again!")
        return None

    except sr.RequestError:
        assistant_speaks("Sorry, there was an error with the speech recognition service.")
        return None

def process_text(text):
    if "calculate" in text:
        # Remove "calculate" from the query and send the rest to WolframAlpha
        query = text.replace("calculate", "").strip()
        try:
            res = client.query(query)
            answer = next(res.results).text
            assistant_speaks(answer)
        except Exception as e:
            assistant_speaks("Sorry, I couldn't process your calculation.")
            print(f"Error: {e}")

    elif "open" in text:
        # Example to open a website
        if "website" in text:
            site = text.split("website")[-1].strip()
            if site:
                url = f"https://{site}"
                webbrowser.open(url)
                assistant_speaks(f"Opening {site}")
            else:
                assistant_speaks("Please provide a valid website name.")

    else:
        # Handle general queries or unknown commands
        assistant_speaks("Sorry, I didn't understand that command.")

# Driver Code
if __name__ == "__main__":
    assistant_speaks("What's your name, Human?")
    name = get_audio()
    if name:
        assistant_speaks("Hello, " + name + '.')

        while True:
            assistant_speaks("What can I do for you?")
            text = get_audio()
            if text:
                text = text.lower()

                if "exit" in text or "bye" in text or "sleep" in text:
                    assistant_speaks("Ok bye, " + name + '.')
                    break

                # Call the process_text function to handle the query
                process_text(text)
```

```python
import subprocess
import os
import speech_recognition as sr
from gtts import gTTS
import playsound
from selenium import webdriver
import random
import wolframalpha
import datetime
import wikipedia
import webbrowser
import requests
import shutil

# Initialize the Wolfram Alpha client
app_id = "YOUR_WOLFRAMALPHA_APP_ID"
wolfram_client = wolframalpha.Client(app_id)

# Function to convert text to speech and play it
def assistant_speaks(output):
    global num
    num = random.randint(1, 10000)  # Random number to avoid file name conflicts
    print("Assistant: ", output)

    tts = gTTS(text=output, lang='en', slow=False)
    file = f"{num}.mp3"
    tts.save(file)
    playsound.playsound(file, True)
    os.remove(file)

# Function to get audio input from the user
def get_audio():
    rObject = sr.Recognizer()
    audio = ''

    with sr.Microphone() as source:
        print("Speak...")
        audio = rObject.listen(source, phrase_time_limit=5)
    print("Stop.")

    try:
        text = rObject.recognize_google(audio, language='en-US')
        print("You: ", text)
        return text.lower()
    except sr.UnknownValueError:
        assistant_speaks("Could not understand your audio, please try again!")
        return ""
    except sr.RequestError:
        assistant_speaks("Sorry, there was an error with the service.")
        return ""

# Function to search the web based on user input
def search_web(input):
    driver = webdriver.Firefox()
    driver.implicitly_wait(1)
    driver.maximize_window()

    input_lower = input.lower()

    if 'youtube' in input_lower:
        assistant_speaks("Opening YouTube")
        indx = input_lower.split().index('youtube')
        query = input.split()[indx + 1:]
        search_query = '+'.join(query)
        driver.get(f"http://www.youtube.com/results?search_query={search_query}")

    elif 'wikipedia' in input_lower:
        assistant_speaks("Opening Wikipedia")
        indx = input_lower.split().index('wikipedia')
        query = input.split()[indx + 1:]
        search_query = '_'.join(query)
        driver.get(f"https://en.wikipedia.org/wiki/{search_query}")

    else:
        if 'google' in input_lower:
            indx = input_lower.split().index('google')
            query = input.split()[indx + 1:]
            search_query = '+'.join(query)
            driver.get(f"https://www.google.com/search?q={search_query}")

        elif 'search' in input_lower:
            indx = input_lower.split().index('search')
            query = input.split()[indx + 1:]
            search_query = '+'.join(query)
            driver.get(f"https://www.google.com/search?q={search_query}")

        else:
            search_query = '+'.join(input.split())
            driver.get(f"https://www.google.com/search?q={search_query}")

# Function to open applications based on user input
def open_application(input):
    input_lower = input.lower()

    if "chrome" in input_lower:
        assistant_speaks("Opening Google Chrome")
        os.system('open -a "Google Chrome"')
    
    elif "safari" in input_lower:
        assistant_speaks("Opening Safari")
        os.system('open -a "Safari"')

    elif "whatsapp" in input_lower:
        assistant_speaks("Opening WhatsApp")
        os.system('open -a "WhatsApp"')

    elif "chatgpt" in input_lower:
        assistant_speaks("Opening ChatGPT")
        os.system('open -a "ChatGPT"')

    else:
        assistant_speaks("Application not available")

# Function to process the user's text input
def process_text(input):
    try:
        if 'search' in input or 'play' in input:
            search_web(input)
            return

        elif "who are you" in input or "define yourself" in input:
            speak = '''Hello, I am your personal assistant.
            I am here to make your life easier. You can command me to perform
            various tasks such as calculating sums or opening applications.'''
            assistant_speaks(speak)
            return

        elif "who made you" in input or "created you" in input:
            speak = "I have been created by Your Name."
            assistant_speaks(speak)
            return

        elif "geeksforgeeks" in input:
            speak = "Geeks for Geeks is the best online coding platform for learning."
            assistant_speaks(speak)
            return

        elif "calculate" in input:
            indx = input.lower().split().index('calculate')
            query = input.split()[indx + 1:]
            res = wolfram_client.query(' '.join(query))
            answer = next(res.results).text
            assistant_speaks("The answer is " + answer)
            return

        elif 'open' in input:
            open_application(input)
            return

        else:
            assistant_speaks("I can search the web for you. Do you want to continue?")
            ans = get_audio()
            if 'yes' in ans or 'yeah' in ans:
                search_web(input)
            else:
                return

    except Exception as e:
        assistant_speaks("I don't understand. I can search the web for you. Do you want to continue?")
        ans = get_audio()
        if 'yes' in ans or 'yeah' in ans:
            search_web(input)

# Main execution code
if __name__ == "__main__":
    assistant_speaks("What's your name?")
    name = get_audio()
    if not name:
        name = "Human"
    assistant_speaks(f"Hello, {name}.")

    while True:
        assistant_speaks("What can I do for you?")
        text = get_audio()

        if text == "":
            continue

        if "exit" in text or "bye" in text or "sleep" in text:
            assistant_speaks(f"Ok bye, {name}.")
            break

        process_text(text)
```
